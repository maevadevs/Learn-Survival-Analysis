# The Cox Proportional Hazards Model

- Learn about the proportional hazards assumption
- Learn the role it plays in fitting and interpreting the *Cox Proportional Hazards* model
- Learn how to predict new subjects' survival times using the *Cox PH* model

## Hazard Function and Hazard Rate

The Cox PH is another model for survival regression

- **Hazard Function $h(t)$** - Describes the probability that the *Event* happens at some time, given survival up to that time
- **Hazard Rate** - The instantaneous rate of the *Event* occurring

We have seen the hazard curve in the *Weibull* model

- The **Hazard Function $h(t)$** and the **Survival Function S(t)** can be derived from each other

$$h(t)=-\frac{d}{dt}logS(t)$$

### Proportional Hazard Assumption

We assume that all individuals' hazards are proportional to one another

- For indivuals $A$ and $B$, it means that

$$h_A(t) = c \cdot h_B(t)$$

where $c$ is some constant

<img src="images/proportional-haard-assumption.png" width=40%>

### Implications of this assumption

There are 2 major implications:

- There is a **Baseline Hazard Function** and other hazard functions are specified with **scalling factors**
- The **Relative Survival Impact** associated with a **variable** does not change with time (time-invariant)

## The Cox Proportional Hazards Model

Based on top of the *Proportional Hazard Assumption*, we have:

$$h(t|x)=b_0(t) \cdot exp(\sum_{i=1}^{n}{b_i(x_i - \bar{x_i})})$$

where:

- $b_0$ is the population-level baseline hazard function that changes with time
- $exp(\sum_{i=1}^{n}{b_i(x_i - \bar{x_i})})$ is the linear relationship between covariates and the log of hazard, does **not** change with time

The **Cox PH** model is another way to run survival regression

- It regresses **covariates** on time-to-event/duration
- It quantifies influences of factors on survival

## Data Requirements for Applying Cox PH Model

- **Durations**: The lifetime/duration of the individuals
- **Events**: Whether the event has been observed (1=Yes, 0=No (censored))
  - If not supplied, the model assumes that no subjects are censored (all hit the event after the duration)
- **Covariates**: Continuous or one-hot-encoded categorical for the regression

## Fitting the Cox PH Model on `echocardiogram`

````python
# Import the CoxPHFitter class
from lifelines import CoxPHFitter
from sklearn.impute import SimpleImputer
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
# Our dataset
echos = pd.read_csv("data/echocardiogram.csv")

display(echos.head())
display(echos.shape)
````

````python
# Checking Missing Values
print(echos.isnull().sum())
print(echos.shape)
````

- We only have 133 observations, but there are many missing values across variables
- Since we don't wanna drop these data, we are gonna implement imputation for missing values with means of each columns

````python
# Impute missing values with mean
imp_mean = SimpleImputer(
    missing_values = np.nan, 
    strategy = 'mean'
)
cols = ['age', 'pericardialeffusion', 'fractionalshortening', 'epss', 'lvdd', 'wallmotion-score']
X = imp_mean.fit_transform(echos[cols])
echos_X = pd.DataFrame(X, columns = cols)
echos_X.shape
````

````python
COLUMNS_keep = ['survival', 'alive']
echos_keep = echos[COLUMNS_keep]
echos_keep.shape
````

````python
echos = pd.concat([echos_keep, echos_X], axis = 1)
echos = echos.dropna() 
display(echos.isnull().sum())
display(echos.shape)
````

- `dropna` function applies to survival and alive variables which we don't consider imputation for
- The number of the data for analysis is 130 observations

````python
# Scatter plots between survival and covariates
sns.pairplot(echos);
````

- For `alive = 1` patients, because they are alive during data collection period and we do not know their survival months after the data collection, they are regarded as censored data
- Hence, the following analysis needs to consider the censored data by making `dead` variable below.

````python
# Feature engineering censorship with `dead` column
echos.loc[echos.alive == 1, 'dead'] = 0
echos.loc[echos.alive == 0, 'dead'] = 1
echos.groupby('dead').count()
````

We have 88 non-censored data and 42 censored data

````python
# Remove `alive` column for now
echos = echos.drop(columns=["alive"])

display(echos.head())
display(echos.shape)
````

Here:

- Duration Column: `survival`
- Event Column (Censorship): `dead`
- Numerical Covariates: `age`, `pericardialeffusion`, `fractionalshortening`, `epss`, `lvdd`, `wallmotion-score`
- Categorical:

````python
# Instantiate the fitter
coxph = CoxPHFitter()
````

The `coxph.fit()` method will need:

- `df`: The dataframe of the time-to-event data
- `duration_col`: The name of the duration column as a string
- `event_col`: The name of the column that contains the censorship, as a string

All columns except `duration_col` and `event_col` are used as covariates

````python
# Fit on the data: This is a right-censored dataset
coxph.fit(
    df=echos,
    duration_col="survival",
    event_col="dead"  # Censorship
)
````

We can access properties from the fitter model, such as:

- `coxph.summary()`
- `coxph.predict()`

**Note: Sometimes, we might want to specify which covariates to use**

- We could remove the unneeded columns from the dataframe
- We could use the formula parameters for `.fit()` and passing the covariates we want to use
  - More convenient
  - But does not scale to large numbers of covariates

```python
coxph.fit(
    df=echos,
    duration_col="survival",
    event_col="dead",  # Censorship
    formula="survival + age * wallmotion-score"
)
```

### Interpreting the Coefficients

We can print the model summary to find the coefficients of the model

````python
# Get the summary of the model
display(coxph.summary)
````

**Hazard Ratio** is the column `exp(coef)`: $e^{coef}$

- It indicates how much the baseline hazard changes with a 1-unit change in the covariate
- Example:
  - `epss` has a coefficient of `0.015`
  - If `epss` increases by `1` => The hazard changes by a factor of $e^{0.015}$, i.e. a 1.5% increase compared to the baseline

## Fitting the Cox PH Model on `prison`

- The DataFrame prison contains information from 432 convicts who were released and followed up for one year after release
- We have previously modeled their time-to-arrest and studied what factors increase or decrease their risk of re-arrest using the Weibull AFT model

````python
# Our dataset
prison = pd.read_csv("data/prison-recidivism.csv")
prison = prison.drop(columns=["index"])
display(prison.head())
display(prison.shape)
````

Here:

- Duration Column: `week`
- Event Column (Censorship): `arrest`
- Numerical Covariates: `fin`, `age`, `race`, `wexp`, `mar`, `paro`, `prio`
- Categorical:

````python
# Instantiate the fitter
coxph = CoxPHFitter()
````

````python
# Fit on the data: This is a right-censored dataset
coxph.fit(
    df=prison, 
    duration_col="week", 
    event_col="arrest"
)
````

````python
# Print model summary
display(coxph.summary)
````

**0.01, 0.05, and 0.1 are common thresholds to use for interpreting significance. Based on the model summary in the console, which sets of covariates are statistically significant at the 0.05 level?**

Looking at the `p` column, we see that the ones with `p < 0.05` are:

- `fin`
- `age`
- `prio`

This means there is a strong correlation between the changes in `fin`, `age` and `prio` and the hazards

### Running a Custom Model

We have isolated 3 factors that are statistically significant at the 0.05 level: `fin`, `age`, and `prio`

- `fin`: if the convict received financial assistance, hazards decrease by 32% (`0.684257 - 1`)
- `age`: for each age older than average, hazards decrease by 6% (`0.944181 - 1`)
- `prio`: for each prior arrest more than average, hazards increase by 10% (`1.095814 - 1`)

Let's now build a custom Cox PH Model using these *Statistically significant* covariates

````python
# Instantiate the fitter
custom_cph = CoxPHFitter()
````

````python
# Fit on the data: This is a right-censored dataset
custom_cph.fit(
    df=prison, 
    duration_col="week", 
    event_col="arrest",
    formula="fin + age + prio"
)
````

````python
# Print model summary
display(custom_cph.summary)
````

**Note**

- When there are many covariates, the problem of overfitting and multicollinearity might also exist for survival regression models
- Read [here](https://lifelines.readthedocs.io/en/latest/fitters/regression/CoxPHFitter.html#lifelines.fitters.coxph_fitter.ParametricSplinePHFitter.fit) for more advanced regularization techniques for CoxPHFitter

## Interpreting the Cox PH Model

**Hazard Ratio**: indicates how much the *baseline* hazard changes with a 1-unit change in the covariate

- What is the *baseline*?
- **Baseline Hazard**: The risk for individuals at the baseline levels for covariates
  - It means setting all the covariates to their *Mean* (*Median* for `lifelines`)`

### Visualizing the Baseline Hazard and Baseline Survival Functions

We use the `.baseline_hazard_` property for accessing the baseline hazard

````python
# Our dataset
prison = pd.read_csv("data/prison-recidivism.csv")
prison = prison.drop(columns=["index"])
display(prison.head())
display(prison.shape)
````

````python
# Instantiate the fitter
coxph = CoxPHFitter()
````

````python
# Fit on the data: This is a right-censored dataset
coxph.fit(
    df=prison, 
    duration_col="week", 
    event_col="arrest"
)
````

````python
# Access the baseline hazard function and plot
coxph.baseline_hazard_.plot()
plt.show()
````

We can see that the baseline hazard varies with time

- Hazard is an instantaneous risk of events
- We see that at $t=20$, there is $1.1%$ chance that an individual will experience an event, given they survive up to $t=20$

We can also use the `baseline_survival_` property to get the survival rates and plot them

- This is the survival curve when all covariates are set to their medians

````python
# Access the baseline survival function and plot
coxph.baseline_survival_.plot()
plt.show()
````

Because the covariates' effects are relative to the baseline, it is important to evaluate them in the context of baseline hazard and survival rates

- The same percent change in hazard at different times may equate different absolute change in hazard

### Interpreting the Hazard Ratio

**Hazard Ratio $e^{coef}$** shows how much hazard changes relative to the *average* individual when covariates change

### Using the `prison` dataset

**What is the impact on survival time from changes in the covariates?**

Fit the Cox PH model to the prison data and calculate the impact on survival time from a one unit increase in `prio` (number of prior arrests)

````python
# Our dataset
prison = pd.read_csv("data/prison-recidivism.csv")
prison = prison.drop(columns=["index"])
display(prison.head())
display(prison.shape)
````

````python
# Instantiate the fitter
coxph = CoxPHFitter()
````

````python
# Fit on the data: This is a right-censored dataset
coxph.fit(
    df=prison, 
    duration_col="week", 
    event_col="arrest"
)
````

````python
# Get the model summary
summary_df = coxph.summary

# Create new column of survival time ratios
summary_df["surv_ratio"] = 1 / summary_df["exp(coef)"]

# Print surv_ratio for prio
print(summary_df.loc["prio", "surv_ratio"])
````

````python
summary_df
````

**With one unit increase in `prio`, Time-to-arrest decrease by 8.7%**

- The ratio is below 1, meaning that survival time decreases as the covariate increases
- The ratio indicates the factor by which the average survival time should be multiplied
- Subtracting 1 from the ratio gives you how much the survival time changes by

$$\frac{1}{0.912564}-1=-0.087436$$

````python
# Comparing survival for 2 convicts
# Name | fin | age | wexp | mar | paro | prio
# Jack |  0  | 35  |  0   |  1  |   1  |  3
# Marie|  1  | 22  |  0   |  0  |  0   |  0

# Plot partial effects
coxph.plot_partial_effects_on_outcome(
    covariates=["fin", "age", "wexp", "mar", "paro", "prio"],
    values= [
        (0,35,0,1,1,3),
        (1,22,0,0,0,0)
    ]
)

# Show plot
plt.show()
````

**Marie is more likely to be arrested sooner because her survival probabilities are lower than those of Jack's**

## The Proportional Hazards Assumption

PHA assumes that that all individuals' hazards are proportional to one another over time

Violating Model assumptions when fitting models often result in sub-optimal analysis

- How do we check the proportional hazard assumptions?
- How important is it?

The PHA means that hazard ratios are time-invariant

- If the covariate only has a few value, inspect each group's K-M curve
- Curves intersect? **Fails the proportional hazard assumption**
  - Covariate value has different effect on survival depending on time
  - The assumption is violated

<img src="images/km-for-pha-assumption.png" width=40%>

- Curves have different shapes? **Fails the proportional hazard assumption**
  - Covariate value has different effect on survival depending on time
  - The assumption is violated
  
<img src="images/km-for-pha-assumption-2.png" width=40%>  

- Curves have similar shapes and are closely paralled? **The assumption is satisfied**

<img src="images/km-for-pha-assumption-3.png" width=40%>

We can check the PHA assumption using **`.check_assumptions()`** method

- We can check if the PHA assumption holds by covariate pairs
- When K-M cannot be used, this is a comprehensive method
- It needs only 1 required parameter: `training_df`
  - The original DataFrame used in the `.fit()` call to fit the model
- `p_value_threshold`
  - The threshold to use to alert the user of violations
  - Default: 0.01
  - Recommended: 0.05

### Using the K-M Curve Approach

````python
# Our dataset
prison = pd.read_csv("data/prison-recidivism.csv")
prison = prison.drop(columns=["index"])
display(prison.head())
display(prison.shape)
````

````python
# Instantiate the fitter
coxph = CoxPHFitter()
````

````python
# Fit on the data: This is a right-censored dataset
coxph.fit(
    df=prison, 
    duration_col="week", 
    event_col="arrest"
)
````

````python
# Testing `fin` using Kaplan-Meier curves
# Import the KaplanMeierFitter
from lifelines import KaplanMeierFitter

# Instantiate a KaplanMeierFitter class
kmf = KaplanMeierFitter()

# Fit kmf to the data of convicts that did not receive financial assistance (fin: 0)
kmf.fit(
    durations=prison[prison["fin"] == 0]["week"],           # Duration
    event_observed=prison[prison["fin"] == 0]["arrest"]     # Censorship
)

# Plot survival curve for fin == 0
ax = kmf.plot()

# Fit kmf to the data of convicts that did receive financial assistance (fin: 1)
kmf.fit(
    durations=prison[prison["fin"] == 1]["week"],           # Duration
    event_observed=prison[prison["fin"] == 1]["arrest"]     # Censorship
)

# Add Plot survival curve for fin == 0
kmf.plot(ax=ax)
plt.show();
````

**Based on the plotted Kaplan-Meier curves, does the covariate fin satisfy the proportional hazards assumption?**

Yes, despite being a visual and seemingly relaxed check, Kaplan-Meier curves that seem parallel indicate the proportional hazards assumption is satisfied

### Using the `check_assumptions()` Approach

````python
# Check if PHA assumption is held
coxph.check_assumptions(
    training_df=prison,
    p_value_threshold=0.05
)
````

````python
# What if we test it on p = 0.1?
coxph.check_assumptions(
    training_df=prison, 
    p_value_threshold=0.1
)
````

**As the output suggests, the proportional hazards assumption test using `.check_assumptions()` can be overly strict and there are [many ways](https://lifelines.readthedocs.io/en/latest/jupyter_notebooks/Proportional%20hazard%20assumption.html) to mitigate the problem.**

### When the Proportional Hazard Assumption Fails

Usually, it's a reasonable assumption

- Violations do not impact model performance significantly
- When it fails, it's best practice to try out other modeling framework and compare AIC scores
  - Weibull AFT
  - Tryout other models that do not have assumptions such as PHA

## Predicting With the Cox PH Model

Using the Cox PH Model, we can analyze existing data and predict survival times for new data

- We can predict after running `.fit()` on the data
- `.predict_median()` Predicts the median lifetime for new subjects
  - If the survival curve does not cross 0.5, the median survival time is `inf`

The `predict_median()` takes 2 parameters:

- `X` - The new dataframe to predict on
- `conditional_after` - A list of values that represent how long subjects have already lived for
  - If defined, all predictions made are conditioned on subjects already having survived for the specified time

It returns a Pandas series for each subject in `X`

### Predicting the Survival Function

`.predict_survival_function` - Predict the survival function of the new data, given their covariates

- Takes the same parameter as `predict_median()`

It returns a Pandas Dataframe of prediction for each subject in `X`

- Each row is a one-time point of the survival function
- Each column is a subject

If `conditional_after` is specified, the time starts at the last observed time of each subject

- The new timeline is the remaining duration of each subject

### Why are survival prediction useful?

- Proactive failure prevention
- Forecasting models

### Key Steps for Right Predictive Modeling

1. Preprocess the data and one-hot encode all categorical variables
1. `train_test_split` the data (Commonly 80-20)
  - The proportion of censored data should be similar in both sets
1. Fit a Cox PH model to the training data
1. Predict on the testing data
1. Measure performance

### Predicting on the `employee_attrition`

- Acquiring new employees as replacements incurs hiring costs and training costs
- We want to predict how long current employees will stay
- `YearsAtCompany` means the duration employees have worked
- `Attrition` indicates if the employee has churned (`1` if churn, `0` otherwise)

````python
# Import libraies
from lifelines import CoxPHFitter
from sklearn.impute import SimpleImputer
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
# Our dataset
employees = pd.read_csv("data/employee-attrition.csv")

# Only keep these columns for now
employees = employees[[
    "Attrition", 
    "YearsAtCompany", 
    "EnvironmentSatisfaction", 
    "JobSatisfaction",
    "RelationshipSatisfaction",
    "WorkLifeBalance",
    "DistanceFromHome",
    "PerformanceRating",
    "HourlyRate",
    "StockOptionLevel"
]]
# Convert Attrition to 0 and 1
employees["Attrition"] = employees["Attrition"].apply(lambda x: 1 if x == "Yes" else 0)

display(employees.head())
display(employees.shape)
````

````python
# Instantiate the fitter
coxph = CoxPHFitter()
````

````python
# Fit coxph on all covariates
coxph.fit(
    df=employees, 
    duration_col="YearsAtCompany", 
    event_col="Attrition"
)
````

````python
# Create a new DataFrame current_employees for all employees that are still present (have not churned)
current_employees = employees.loc[employees["Attrition"] == 0]

# Existing durations of employees that have not churned
current_employees_last_obs = current_employees["YearsAtCompany"]
````

Now, we can predict when present employees might churn based on the fitted survival function and the variables we created

````python
# Predict survival function conditional on existing durations
coxph.predict_survival_function(
    X=current_employees,
    conditional_after=current_employees_last_obs # given how long they have already worked for
)
````

````python
# Predict median remaining times for current employees
pred_median = coxph.predict_median(
    X=current_employees,
    conditional_after=current_employees_last_obs # given how long they have already worked for
)

# Print the predicted remaining times
print(pred_median)
````
